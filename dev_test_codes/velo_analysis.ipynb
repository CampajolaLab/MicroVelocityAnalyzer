```python
with open("data/stablecoins/mvresults/USDT_balances_34080417.pickle", 'rb') as handle:
    bal = pickle.load(handle)
```

```python
def normalize_vel(vel, bal):
    return np.array([vel[i]/bal[i] if bal[i] > 0 else 0 for i in range(len(bal))])
```

```python
norm_vel = {i:normalize_vel(vel[i], bal[i]) for i in vel.keys()}
```
